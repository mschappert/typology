```python
#chunk = {'band': 1, 'x': 1024, 'y': 1024}

def img_diff(image1, image2, output_dir, metric = metric): # chunks=chunk
    """
    Subtract two TIFF rasters using rioxarray + Dask, preserving metadata
    and allowing negative values.

    Parameters:
    image1_path (str): Path to the earlier image (e.g., 1990).
    image2_path (str): Path to the later image (e.g., 1995).
    output_dir (str): Path where the difference raster will be saved.
    chunks (dict): Dask chunk sizes by dimension.
    """
    
    # Open each raster lazily with Dask
    img1 = rxr.open_rasterio(image1, chunks={'band': 1, 'x': 1024, 'y': 1024}) # chunks=chunks
    img2 = rxr.open_rasterio(image2, chunks={'band': 1, 'x': 1024, 'y': 1024}) # chunks=chunks

    # Align dimensions and coordinates if needed
    #img2, img1 = xr.align(img2, img1, join='exact')

    # subtract images (later minus earlier)
    diff = img2 - img1

    # Preserve CRS and transform, then write with Dask in parallel
    diff.rio.write_crs(img1.rio.crs, inplace=True)
    diff.rio.write_transform(img1.rio.transform(), inplace=True)
    
    # create naming convention
    base1 = os.path.splitext(os.path.basename(image1))[0]
    base2 = os.path.splitext(os.path.basename(image2))[0]
    # pull year from begining of name (ie 1990)
    year1 = ''.join(filter(str.isdigit, base1))[:4]
    year2 = ''.join(filter(str.isdigit, base2))[:4]
    # take just last 2 digits (ie 90)
    label1 = year1[-2:]
    label2 = year2[-2:]
    
    # output file name "90-95_area.tif"
    out_name = f"{label1}-{label2}_{metric}.tif"
    out_path = os.path.join(output_dir, out_name)

    # Write raster using Dask
    diff.rio.to_raster(
        out_path,
        dtype='int32',
        compress='LZW',
        tiled=True #,
        #num_workers='all_cpus',  # use all available cores
        #lock=True
    )

    return output_dir
```

```python
if __name__ == "__main__":
    result = img_diff(image1=earlier, image2=later, output_dir=output, metric=metric) #chunks=chunk
    print(f"Saved difference raster to {result}")
```
